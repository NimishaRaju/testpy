```python
class Aircraft:
    def __init__(self, code, units, manu, dist, occurances = 1):
        self.remainingOccurances = occurances
        self.rangeModifier = 1
        self.setCode(code)
        self.setUnits(units)
        self.setManufacturer(manu)
        self.setFlightRange(dist)
        self.route = None
       
    def printDetails(self):
        print(str(self.getCode()) + " <--> " + str(self.getManufacturer()) + " <--> " + str(self.getUnits()) + " <--> " + str(self.getFlightRange()))
    
    def isManufacturedBy(self, manufacturer):
        return self.getManufacturer() == manufacturer
    
    def increaseOccurances(self, amount = 1):
        self.remainingOccurances += amount
        
    def decreaseOccurances(self):
        self.remainingOccurances -= 1
        if self.remainingOccurances <= 0:
            AircraftAtlas.removeAircraft(self.getCode())
    
    def setCode(self, code):
        self.code = code
        
    def setUnits(self, units):
        self.units = units
        #print(self.units)
        #if the units were given in miles, adjust the range modifier
        if(self.units == "imperial"):
            self.rangeModifier = 1.6093
        
    def setManufacturer(self, manu):
        self.manufacturer = manu
        
    def setFlightRange(self, dist):
        self.flightRange = dist * self.rangeModifier
        
    def getCode(self):
        return self.code
        
    def getUnits(self):
        return self.units
        
    def getManufacturer(self):
        return self.manufacturer
        
    def getFlightRange(self):
        return self.flightRange
    
    #instantiate a Route object with the airport codes to be visited by this aircraft
    def setRoute(self, routeNames):
        self.route = Route(routeNames)
        self.decreaseOccurances()
```

```python
#holds the different airports needed to be reached for an aircraft and performs calculations on distances and costs
class Route:
    def __init__(self, airportCodes):
        self.legs = [] #list of airports for the route
        self.airports = []
        self.setAirports(airportCodes)
        self.printRoute()
        self.distTable = self.generateDistanceTable()
        self.costTable = self.generateCostTable()
        self.generatePermutations()
        
    #loop through the airport codes provided, fetch the airports from the atlas and append them to the list for reference
    def setAirports(self, airportCodes):
        #fetch airports from atlas
        #self.homeAirport = AirportAtlas.getAirport(airportCodes[0])
        for code in airportCodes:
            self.legs.append(AirportAtlas.getAirport(code))
            self.airports.append(AirportAtlas.getAirportName(code))
            #self.legs = [AirportAtlas.getAirport(airportCodes[0]), AirportAtlas.getAirport(airportCodes[1]), AirportAtlas.getAirport(airportCodes[2]), AirportAtlas.getAirport(airportCodes[3]), AirportAtlas.getAirport(airportCodes[4])]
        
    def printRoute(self):
        #self.homeAirport.printDetails()
        for leg in self.legs:
            leg.printDetails()
        print()
        
    def generateDistanceTable(self):
        distances = {}

        for src_airport in self.legs:
            distances[src_airport.getName()] = {}
            for dest_airport in self.legs:
                distances[src_airport.getName()][dest_airport.getName()] = self.greatCircleDistance2(src_airport, dest_airport)

        print(distances)
        print()
        return distances
        
    def generateCostTable(self):
        # calculate journey costs and return dictionary of dictionaries
        journeyCosts = []

        for i in range(len(self.legs)):
            #journeyCosts[src_airport.getName()] = 
            jc=[]
            for j in range(len(self.legs)):
                jc.append(self.journeyCost2(self.legs[i],self.legs[j]))
            journeyCosts.append(jc)

        #print(journeyCosts)
        return journeyCosts
        
    def greatCircleDistance2(self, source, destination):
        '''Function to calculate the distance between to points using the 
        great circle. Function arguments to be passed as coordinate list
        Location coordinates stored in location dictionary.
        Function to return distance in km to 2 decimal places'''
    
        # mean earth radius (earth is a flattened sphere)
        radius = 6371

        # set coordinates
        x1 = source.getLat()
        y1 = source.getLng()
        x2 = destination.getLat()
        y2 = destination.getLng()

        # convert to radians for centralAngle formula
        x1 = math.radians(x1)
        y1 = math.radians(y1)
        x2 = math.radians(x2)
        y2 = math.radians(y2)

        # angle between the two airports in radians
        a = math.sin(x1) * math.sin(x2) + math.cos(x1) * math.cos(x2) * math.cos(y1 - y2)
        # round to avoid floating point error
        centralAngle = math.acos(round(a,5))

        # arc length = radius * angle(in radians)
        arcLength = radius * centralAngle

        return round(arcLength, 2)
    
    def journeyCost2(self, source, destination):
        '''Function to calculate the journey costs between to points using the 
        great circle function. Fuel cost based on departure airport 
        Function to return cost in EUR to 2 decimal places'''
        return round(self.greatCircleDistance2(source, destination) * source.getCurrencyMultiplier(), 2)
    
    #using graphhhhhhhhhhhhhhhhhhhhhhhhh
    
    def generatePermutations(self):
        print("printing my valus")
        self.vertices=self.airports
        print(self.vertices)
        self.edges=self.generateCostTable()
        print(self.edges)
        self.routeair=list(range(len(self.legs)))
        start_end_airport = self.routeair.pop(0)
        permu = list(itertools.permutations(self.routeair))
        totalCost=0
        lowestCost=list()
        routelist=list()
        for val in permu:
            val= (start_end_airport,)+val+(start_end_airport,)
             #print(n)
            for i in range(len(val)-1):
                cur=val[i]
                nex=val[i+1]
                totalCost+=self.edges[cur][nex]
            lowestCost.append(totalCost)
            if len(lowestCost)>=2:
                print(lowestCost)
                if lowestCost[0]>lowestCost[1]:
                    lowestCost[0]=lowestCost[1]
                    routelist=val
                lowestCost.pop(1)
            totalCost=0
        print(lowestCost[0])
        for val in routelist:
            print(self.vertices[val],"<--->",end="")
        
        
        
        
        print("ending---------------------------------------------------------------------")
```

```python
class Row:
    #constructor is passed the row heading, the list of headings(columns) and the type of data the table is holding
    def __init__(self, heading, headingsList, dataType = None):
        self.heading = heading
        self.headingsList = headingsList
        self.cells = []
        self.createCells(dataType)
        
    #create each cell in the row
    def createCells(self, dataType):
        for heading in self.headingsList:
            if dataType == "distance":
                cell = Cell(heading, self.heading, greatCircleDistance2(locations[self.heading], locations[heading]))
            elif dataType == "cost":
                cell = Cell(heading, self.heading, round(greatCircleDistance2(locations[self.heading], locations[heading]) * currency[self.heading], 2))
            else:
                cell = Cell(heading, self.heading, (self.heading[0] + "->" + heading[0]))
            self.cells.append(cell)
            
    def printRow(self):
        print(self.heading + ":   ", end = "")
        
        for cell in self.cells:
            print("\t" + str(cell.getData()), end="")
            
        print()
        
```

```python
class Table:
    def __init__(self, headingsList, tableType = None):
        self.headingsList = headingsList
        self.headingsMapping = {}      #dictionary to map the row objects to their row names, for easier access
        self.rows = []
        self.tableType = tableType
        self.createRows()
    
    def createRows(self):
        for heading in self.headingsList:
            row = Row(heading, self.headingsList, self.tableType)
            self.addRow(row, heading)
    
    def addRow(self, row, heading):
        self.rows.append(row)
        self.headingsMapping[heading] = row
        
    def getRow(self, heading):
        return self.headingsMapping[heading]
        
    def printTable(self):
        print("\t", end = "")
        for col in self.headingsList:
            print("\t" + col, end = "")
            
        print()
        
        for row in self.rows:
            row.printRow()
        
```
