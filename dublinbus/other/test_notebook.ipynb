```python
def testing_with_stops_times_csv():

    host='127.0.0.1'
    user='postgres'
    password='23081987'
    db='postgres'
    
    sql = """SELECT 
                    trip_id,
                    stop_id,
                    stop_sequence,
                    stop_headsign,
                    arrival_time,
                    departure_time 
                    
                    from dbus_data.stop_times"""
    
    db = psycopg2.connect(host=host,database=db, user=user, password=password)

    cursor = db.cursor()
    cursor.execute(sql)
    result = cursor.fetchall()
    db.close()
    
    #print(result)
    return result
    #result = json.dumps(result)
```

### Find bus number in trip_id, append it at the end of each row

```python
i=0
new_data_list = [0]*len(result)
for data in result:
    trip_id = data[0]
    x = trip_id.find("-")
    y = trip_id.find("-",x+1)
    bus_number = trip_id[x+1:y]
    new_data_list[i] = list(result[i])
    new_data_list[i].append(bus_number)      
    i+=1

print(new_data_list[:5])
```

### Get unique stops

```python
used = set()
unique = [x for x in new_data_list if x[1]+str(x[2])+x[3]+x[6] not in used and ((used.add(x[1]+str(x[2])+x[3]+x[6])) or True)]
```
